```python
from IPython.core.pylabtools import figsize
import seaborn as sns
sns.set()
from matplotlib.patches import PathPatch

df2 = pd.read_excel(r'https://github.com/maurorodas/414G7G/raw/main/data/data.xlsx', sheet_name='Temperatures')

df2.boxplot(by='YEAR',color='blue',figsize=(20,12),fontsize=15,rot=90)
```

```python
import seaborn as sns
dd=pd.melt(df2,id_vars=['YEAR'],value_vars=['MANIZALES','BOGOTA','CALI','MEDELLIN'],var_name='Ciudades')
sns.set(rc={'figure.figsize':(30,12)})
sns.boxplot(x='YEAR',y='value',data=dd,hue='Ciudades')
```

```python
dfmeans = df2.groupby('YEAR').mean()

plt.plot(dfmeans, linestyle='--', marker='o')
```
